```python
import re
def Normale_class_format(classe):
    """Convertit les formats de classe variés en un format standardisé (ex: 4emA)."""
    classe = str(classe).lower()  # Convertir en minuscules pour une meilleure correspondance

    # Utiliser des expressions régulières pour extraire le niveau et la lettre
    match = re.search(r"(\d+)\s*(?:em|ie?m)\s*([a-z])", classe)

    if match:
        niveau = match.group(1)
        lettre = match.group(2).upper()
        return f"{niveau}em{lettre}"
    else:
        return classe  # Retourne la classe originale si aucun format reconnu 
# Obtenir l'index de la colonne "Classe"
try:
        classe_index = header.index('Classe')  # Trouver la position de la colonne "Classe"
except ValueError:
        print("La colonne 'Classe' n'a pas été trouvée dans le fichier CSV.")
        # Gérer le cas où la colonne "Classe" est manquante si nécessaire

    # Itérer sur data_dict et mettre à jour les valeurs de "Classe"
for row in data_dict:
        # Obtenir la valeur originale de "Classe" depuis data_dict
        original_classe = row['Classe']
        # Normaliser la valeur de "Classe"
        normalized_classe = Normale_class_format(original_classe)
        # Mettre à jour la valeur de "Classe" dans data_dict
        row['Classe'] = normalized_classe

print(type(data_dict[0]["Classe"]))
data_dict
```

```python
def separateurDate(L):
    L = str(L)
    t = []
    j = 0
    for i in range(len(L)):
        if L[i] in "|/:-_  ":
            t.append(L[j:i])
            j = i + 1
    t.append(L[j:])
            
    
    return t
a = separateurDate("12/05_2000")
a
```

```python
import locale
from datetime import datetime

def est_date_valide(date_str, format_str):
    # Essayer de définir la locale, si cela échoue, utiliser la locale par défaut
    try:
        locale.setlocale(locale.LC_TIME, 'fr_FR.UTF-8')  
    except locale.Error:
        print("Avertissement: la locale 'fr_FR.UTF-8' n'est pas prise en charge, utilisation de la locale par défaut.")
    
    for format in format_str:
        try:
            datetime.strptime(date_str, format)
            return True
        except (ValueError, TypeError):
            pass
    return False

a = est_date_valide("12/février/2002", ["%d/%m/%Y", "%Y-%m-%d", "%d/%B/%Y" , "%d/%b/%Y"])
print(a)  
```

```python
def Numero_valide(Numero):
  if len(Numero) != 7:  # Vérifier la longueur en premier
    return False

  for i in range(len(Numero)):
    # Vérifier si le caractère est valide (chiffres ou 'A' ou 'Z')
    if Numero[i] not in "0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ":
      return False  # Retourner False si un caractère invalide est trouvé

  return True  # Retourner True si tous les caractères sont valides et la longueur est correcte

a = Numero_valide("40DKG6T")
print(a)
```

```python
def Moyenne_etudiant(chaine_note):
    resultat = {}
    moyennes_matieres = []

    chaine_note = str(chaine_note)
    chaine_note_matiere = chaine_note.split("#")
    
    for i in chaine_note_matiere :
        Nom_matiere = i.split("[")[0]
          
        notes_str = i[i.find("[") + 1:i.find("]")]
        
        notes_list = notes_str.replace("|", ":").split(":")

        
        Note_devoirs = []
        Note_examen = 0
        try:
            Note_devoirs = [float(note.replace(',', '.')) for note in notes_list[:-1]]
            Note_examen = float(notes_list[-1])
        except (ValueError, IndexError):
            resultat[Nom_matiere] = 'N/A'
            continue
         # Calculate average if notes are valid
        if all(0 <= note <= 20 for note in Note_devoirs + [Note_examen]):
            moyenne = (sum(Note_devoirs) / len(Note_devoirs) + 2 * Note_examen) / 3
            resultat[Nom_matiere] = moyenne
            moyennes_matieres.append(moyenne) 

            moyenne_General = sum(moyennes_matieres) / len(moyennes_matieres) if moyennes_matieres else 0
        else:
            print("note non valide")
            return False
        
    resultat["moyenne_general"] = moyenne_General
    
    return resultat

a = Moyenne_etudiant("Math[14|2:13]#Francais[12|11|8:13]#Anglais[13,5|11:15]#PC[11:9]#SVT[12|9|13|11:11]#HG[10:13]Francais[4|11:13]#Anglais[13,5:15]#PC[11:9]#SVT[12|9|16|11:12]#HG[10:13]#Math[12|14,5|11:13]")
print(a)   
```

```python
def information_invalide():
    errors = []  # Liste pour stocker les messages d'erreur
    for i in data_dict:
        
        if Numero_valide(i["Numero"]) and Nom_Valide(i["Nom"]) and Prenom_Valide(i["Prénom"]) and est_date_valide(i["Date de naissance"], ["%d/%m/%Y", "%m/%d/%Y", "%Y/%m/%d", "%d/%m/%y", "%d/%B/%Y", "%d%/%b/%Y"]) and ChaineNote_valide(i["Note"]) and classe_valide(i["Classe"]):
            continue  # Passer à l'élément suivant s'il est valide
        else:
            error_msg = ""  # Initialiser le message d'erreur pour cet élément
            if not Numero_valide(i["Numero"]):
                error_msg += f"Le numero '{(i['Numero'])}' est invalide. "
            try:
                if not Nom_Valide(i["Nom"]):
                    error_msg += f"Le Nom '{(i['Nom'])}' est invalide. "
           
                if not Prenom_Valide(i["Prénom"]):
                    error_msg += f"Le Prénom '{i["Prénom"]}' est invalide."
            except(IndexError):
                pass
            if not est_date_valide(i["Date de naissance"], ["%d/%m/%Y", "%m/%d/%Y", "%Y/%m/%d", "%d/%m/%y", "%d/%B/%Y", "%d%/%b/%Y"]):
                error_msg += f"Le Date de naissance'{i["Date de naissance"]}' est invalide."
            if not ChaineNote_valide(i["Note"]):
                error_msg += f"La chaine de  note '{i["Note"]}' est invalide."
            if not classe_valide(i["Classe"]):
                error_msg += f"La Classe '{i["Classe"]}' est invalide."
            errors.append(error_msg)  # Ajouter le message d'erreur à la liste
            #print(f"\033[31m{error_msg}\033[0m") # Afficher le message d'erreur en rouge dans la cellule de sortie

    if errors:  # Si des erreurs ont été trouvées
        print("Le nombre d'elements invalides est : ",len(errors))
        return False, errors  # Renvoyer False et la liste des erreurs
    else:
        return True, []  # Renvoyer True et une liste vide (aucune erreur)
a = information_invalide()
a
```
